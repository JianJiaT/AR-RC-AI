```python
import os
import pandas as pd
import torch
from torch.utils.data import Dataset
from torch.utils.data import DataLoader
import shutil
```

```python
os.chdir("C:\\Users\\tjjtj\\CG4002\\glove_data\\combined_data")
print(os.getcwd())
```

```python
def extract_before(string, substring):
    index = string.find(substring)
    return string[:index]
```

```python
datafolders = []
for filename in os.listdir():
    if ("_datafolder" in filename and filename != "empty_datafolder"):
        datafolders.append(filename)
print(datafolders)

filepaths = []
for folder in datafolders:
    os.chdir(folder)
    if ("split_data" in os.listdir()):
        shutil.rmtree("split_data")
    os.mkdir("split_data")
    for filename in os.listdir():
        if (".csv" in filename):
            with open(filename, "r") as f:
                contents = f.read()
                line_array = []
                lines = contents.strip().split("\n")
                remainder = len(lines) % 100
                rounded = len(lines) - remainder
                lines = lines[50:rounded]
                
                line_count = 0
                split_count = 0
                for line in lines:
                    num_array = []
                    nums = line.strip().split(",")
                    for num in nums:
                        num_array.append(float(num))
                    line_array.append(num_array)
                    line_count += 1
                    if line_count == 50:
                        os.chdir("split_data")
                        new_file_name = extract_before(filename, ".csv") + "_" + str(split_count) + ".csv"
                        df = pd.DataFrame(line_array)
                        df.to_csv(new_file_name, header=False, index=False)
                        line_array = []
                        line_count = 0
                        split_count += 1
                        filepaths.append(folder + "/split_data/" + new_file_name)
                        os.chdir("..")
                
    os.chdir("..")

with open("labels.csv", "w") as f:
    for filepath in filepaths:
        f.write(filepath + ", ")
        if ("stationary" in filepath):
            f.write("0\n")
        elif ("random" in filepath):
            f.write("1\n")
        elif ("wave" in filepath):
            f.write("2\n")
        elif ("punch" in filepath):
            f.write("3\n")
        elif ("swipe" in filepath):
            f.write("4\n")
        elif ("wipe" in filepath):
            f.write("5\n")
        elif ("shake" in filepath):
            f.write("6\n")
        elif ("hammer" in filepath):
            f.write("7\n")
```

```python
gesture_class_dict = {"stationary": "0", "random": "1", "wave": "2", "punch": "3", "swipe": "4", "wipe": "5", "shake": "6", "hammer": "7"}
for gesture in gesture_class_dict:
    folder = gesture + "_datafolder"
    
    os.chdir(folder)
    
    filepaths = []
    os.chdir("split_data")
    for filename in os.listdir():
        filepaths.append(folder + "/split_data/" + filename)
    os.chdir("..")
    os.chdir("..")
    
    csv_name = gesture + "_labels.csv"
    with open (csv_name, "w") as f:
        for filepath in filepaths:
            f.write(filepath + ", ")
            f.write(gesture_class_dict[gesture] + "\n")
```
